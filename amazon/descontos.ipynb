```python
import pandas as pd
```

```python
df = pd.read_csv('./data/dados_Imersão - dados.csv')
df.head()
```

# dados básicos do do arquivo

```python
quantidade_registros = len(df)
print("Registros ", quantidade_registros)
print("colunas:",df.shape[1])
```

```python
# tratando dados
df['product_price'] = df['product_price'].str.replace('₹', '').str.replace('.', '').str.replace(',', '.')
df['product_price'] = df['product_price'].astype(float)

df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace('.', '').str.replace(',', '.')
df['actual_price'] = df['actual_price'].astype(float)
```

```python
df['porcent'] = ((df['product_price'] - df['actual_price']) / df['product_price']) * 100
df.head()
```
