```python
import pandas as pd
from currency_converter import CurrencyConverter
from IPython.display import display
```

```python
df = pd.read_csv('./data/dados_Imersão - dados.csv')
df.head()
```

# dados básicos do arquivo

```python
quantidade_registros = len(df)
print("Registros ", quantidade_registros)
print("colunas:",df.shape[1])
```

```python
# tratando dados
df['product_price'] = df['product_price'].str.replace('₹', '').str.replace('.', '').str.replace(',', '.')
df['product_price'] = df['product_price'].astype(float)

df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace('.', '').str.replace(',', '.')
df['actual_price'] = df['actual_price'].astype(float)

df['rating'] = df['rating'].str.replace(',', '.')
df['rating'] = df['rating'].astype(float)


```

```python
df['porcent'] = ((df['product_price'] - df['actual_price']) / df['product_price']) * 100

# conerter valor para BLR
# cotacao = 0.06
# converter valor para BLR
c = CurrencyConverter()
cotacao = c.convert(1, 'INR', 'BRL')

df['actual_price_BLR'] = df['actual_price'] * cotacao
df.head() 
```

```python
notas_acima_de_4_5 = df[df['rating'] > 4.5]


def highlight_above_4_5(val):    
    color = ''
    if val >= 5.0:
        color = 'background-color:green'
    elif val > 4.8:
        color = 'background-color:lightgreen'
    elif val > 4.7:
        color = 'background-color:limegreen'
    elif val >= 4.5:
        color = 'background-color:palegreen'
    return color

    

df_sorted = df.sort_values(by='rating', ascending=False)
df_styled = df_sorted.style.applymap(highlight_above_4_5, subset=['rating'])
notas_acima_de_4_5 = notas_acima_de_4_5.style.applymap(highlight_above_4_5, subset=['rating'])


df_styled_sorted = df_styled.data.sort_values(by='rating', ascending=False)

# df_styled_sorted
# df_styled
```

# Valores

```python
produtos_abaixo_de_50 = (df['actual_price_BLR'] < 50).sum()
media_abaixo_de_50 = df[df['actual_price_BLR'] < 50]['actual_price_BLR'].mean()
print('Abaixo de R$50:', produtos_abaixo_de_50)
print('média:',media_abaixo_de_50)

produtos_entre_50_e_500 = ((df['actual_price_BLR'] >= 50) & (df['actual_price_BLR'] <= 500) ).sum()
print('entre R$50 e R$500:', produtos_entre_50_e_500)

produtos_acima_de_500 = (df['actual_price_BLR'] > 500).sum()
print('maiores que R$500:', produtos_acima_de_500)


```
