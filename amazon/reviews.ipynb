```python
df1 = pd.read_csv('./data/dados_Imersão - dados.csv')
df2 = pd.read_csv('./data/Avaliacoes.csv')
df = pd.merge(df1, df2, on=['product_id'], how='left')

df.head()
```

```python
# tratando dados
df['product_price'] = df['product_price'].str.replace('₹', '').str.replace('.', '').str.replace(',', '.')
df['product_price'] = df['product_price'].astype(float)

df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace('.', '').str.replace(',', '.')
df['actual_price'] = df['actual_price'].astype(float)

df['rating'] = df['rating'].str.replace(',', '.')
df['rating'] = df['rating'].astype(float)

df.head()
```

```python
quantidade_registros = len(df)
print("Registros ", quantidade_registros)
print("colunas:",df.shape[1])
```

# análise de satisfação

```python
def analisar_sentimento(texto):
    analysis = TextBlob(texto)        
    if analysis.sentiment.polarity > 0:
        return 'Positivo'
    elif analysis.sentiment.polarity < 0:
        return 'Negativo'
    else:
        return 'Neutro'
    
def analisar_sentimento_polarity(texto):
    analysis = TextBlob(texto)    
    return analysis.sentiment.polarity    

def traduzir_para_portugues(titulo):
    blob = TextBlob(str(titulo))
    titulo_traduzido = str(blob.translate(to='pt'))
    return titulo_traduzido



df['sentimento'] = df['review_content'].apply(analisar_sentimento)
df['sentimento_polarity'] = df['review_content'].apply(analisar_sentimento_polarity)



# df['review_title_pt'] = df['review_title'].apply(traduzir_para_portugues)


df.head()
```

```python
contagem_sentimento = df['sentimento'].value_counts()
contagem_sentimento = contagem_sentimento.sort_index(ascending=True)
contagem_sentimento
```
